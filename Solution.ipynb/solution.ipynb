EXPLORATRY DATA ANALYSIS

GENDER


```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
gender_counts = df["Gender"].value_counts()

# Display the counts
print(gender_counts)

# Plotting a bar chart
plt.bar(gender_counts.index, gender_counts)

# Adding labels and title
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Distribution of Gender")

# Display the plot
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame and has been loaded successfully

# Plotting a bar chart for gender vs. total amount without using an estimator
sns.barplot(x='Gender', y='Amount', data=df,estimator=sum,hue='Gender')

# Adding labels and title
plt.xlabel("Gender")
plt.ylabel("Mean Total Amount")
plt.title("Gender vs. Mean Total Amount")

# Display the plot
plt.show()
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame and has been loaded successfully

# Grouping by 'Age Group' and calculating the sum of 'Amount'
df_grouped = df.groupby('Age Group', as_index=False)['Amount'].sum()

# Sorting the DataFrame by 'Amount' in descending order
df_sorted = df_grouped.sort_values(by='Amount', ascending=False)

# Plotting a bar chart for Total Amount vs Age Group
sns.barplot(x='Age Group', y='Amount', data=df_sorted,hue='Age Group')

# Adding labels and title
plt.xlabel("Age Group")
plt.ylabel("Total Amount")
plt.title("Total Amount vs Age Group")

# Display the plot
plt.show()

```

                OCCUPATIONS

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

            PRODUCT CATEGORY

```python
ax = sns.countplot(data = df, x = 'Product_Category', hue='Product_Category')
sns.set({'figure.figsize':(10,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sale_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)
print(sale_state)
sns.barplot(data=sale_state,x='Product_Category',y='Amount',hue='Product_Category')
plt.title("Top 5 Product Category by Amount")
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
Product_ID_orders=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
print(Product_ID_orders)
sns.set({'figure.figsize':(30,5)})
sns.barplot(data=Product_ID_orders,x='Product_ID',y='Orders',hue='Product_ID')
plt.title("Top Ordered product_ID")
```
